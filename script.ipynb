# [dbo].[sp_state]

```sql
/****** Object:  StoredProcedure [dbo].[sp_state]    Script Date: 01-12-2023 10:38:25 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_state]
@state_id int=0,
@state_name varchar(100)=null,
@country_id int=0,
@action int=0
AS
BEGIN
	if(@action=1)
	begin
	insert into tbl_state values (@state_name,@country_id);
	end

	if(@action=2)
	begin
	select tbl_state.*,tbl_country.country_name from tbl_state left join tbl_country on tbl_state.country_id=tbl_country.country_id;
	end

	if(@action=3)
	begin
	select * from tbl_state where state_id=@state_id;
	end

	if(@action=4)
	begin
		update tbl_state set state_name=@state_name where state_id=@state_id; 
	end

	if(@action=5)
	begin
		delete from tbl_state where state_id=@state_id;
	end
	if(@action=6)
	begin
		select * from tbl_state;
	end
END
GO
```
